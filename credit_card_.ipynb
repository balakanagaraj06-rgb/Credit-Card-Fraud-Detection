Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to panda DataFrame
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
# Distribution of legit transaction &  fradulent transaction
credit_card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0--->Normal Transaction
1--->Ffradulant Transaction

```python
# seperating the data for anaiysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

under-sampling


Build a sample dataset containing similsr distribution of normal transaction and fradulant transactions

Number of Fradulent transaction ---> 86

```python
legit_sample = legit.sample(n=86)
```

Concattenating two DataFrames

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Spliting the data into Features & Targets

```python
x = new_dataset.drop(columns='Class',axis=1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Split the data into Training data & Testing Data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the logistic regression model
model.fit(x_train, y_train)
```

Model Evaluation


Accuracy Score

```python
# accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on training data :', training_data_accuracy )
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score on Test Data :', test_data_accuracy)
```
